```python
#cleaning data 
# handling missing values
import pandas as pd
df = pd.read_csv("raw_data.csv")

df.isnull()
df.isna()
df.isnull().sum()

df.dropna()
df.dropna(axis=1)

df.fillna(0)

age_mean = df["age"].mean()
df["age"].fillna(age_mean)

cleaned_data= df.copy()
cleaned_data["age"]=cleaned_data["age"].fillna(age_mean)
cleaned_data

df.ffill()
df.bfill()
```

```python
#feature eng 
df2 = df.copy();
df2["tax"] = df2["income"].apply(lambda x: "20%" if x>=60000 else "10%") #applying tax based on the condition

gender_map={"Male":"M","Female":"F","Unknown":"U"}
df2["gender"]=df2["gender"].map(gender_map) #maping the gender from Female --> F

df2=df2.assign(new_income=df2["income"]*1.1) #assigning new coln with 10 % inc in income

df2["country"]=df2["country"].replace("USA","United States") #replalcing usa --> United States

df2.columns =["ID","Name","Age","Country","Gender","Income","Tax","New_Income"] # renaming the colms from id --> Id
#df2.rename(columns={"Income":"Salary"}) # renamed the particular colmn Income --> Salary
df2.rename(index={1:"first"}) # rename the particular row 1 --> First 

df2.sort_values("Income") # sorting income in ascending ord
df2=df2.fillna(50)
df2.sort_values("Income",ascending =False) #sorting in descending ord
sorted_df =df2.sort_values(["Income","Age"]) # sorting basis of 2 columns (based on income , but if have same income checks age)

#reset
sorted_df.sort_index() #sorting the index (to its prev/starting phase)
sorted_df.reset_index()

#ranking
#sorted_df["Ranking"]=sorted_df["Income"].rank(ascending =False,method ="dense")

#reordering colmns 
sorted_df[["ID","Name","Age","Country","Gender","Income","New_Income","Tax"]]
```

```python
 # melt and pivot
import pandas as pd
df =pd.DataFrame({
    "country":["USA","USA","India","India"],
    "year":[2020,2021,2020,2021],
    "sales":[100,120,90,110],
    "profit":[20,25,18,22],
})

##coverting wide to long -->using melt
melted_df =df.melt( 
    id_vars=["country","year"],
    value_vars=["sales","profit"],
    var_name="metrics",
    value_name="value"
)

##converting long to wide -->using pivot
original_df =melted_df.pivot(
    index=["country","year"],
    columns="metrics",
    values="value"
)
```
