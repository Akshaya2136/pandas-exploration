```python
import pandas as pd
```

```python
info={
    "Name" :["Akshaya","Anand","Mamatha"],
    "CGPA" :[9.0, 9.5 ,9.0]
}
df = pd.DataFrame(info)
print(df)
```

```python
df
```

```python
# creating a series(1d labeled array)
s =pd.Series([1, 2, 3, 4, 5])
print(s)
#index
print(type(s))
print(s[0])
```

```python
s =pd.Series([23, 22, 23, 24, 25],index=["adam","eve","bob","charlie","lara"])
print(s)
print(s["adam"])
print(s["bob"])
print(s.index)
```

```python
#vectorized operation
s1 =pd.Series([1, 2, 3, 4, 5])
s2 =pd.Series([10, 20, 30, 40, 50])
print(s1+s2)
s1[0]=100
changed_s1=s1.drop(0)
print(s1)
print(changed_s1)
```

```python
#Data Frames 
info={
    "Name" : ["Adam", "eve", "bob"],
    "Age" : [23, 24, 25],
    "CGPA" : [9.5, 9.0, 8.5]
}
df =pd.DataFrame(info)
print(df,type(df))
print(df.index)
```

```python
#creating dataframes from list pf lists
df=pd.DataFrame([["adam",23,9.5], ["sam",24,9.0], ["bob",25,8.5]],columns=["Name","Age","CGPA"])
print(df)
```

```python
#creating data frames from numpy
import numpy as np
arr= np.array([[1,2,3],[4,5,6]])
df=pd.DataFrame(arr,columns=["A","B","C"])
print(df)
```
