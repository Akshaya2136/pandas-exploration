```python
# importing the IRIS datset
import pandas as pd
df = pd.read_csv("IRIS.csv")
df.head(10)
print(type(df))
print(df.shape)
df.describe()
```

```python
df
```

```python
df.query("petal_length>4.5")
```

```python
species_var="Iris-virginica"
df.query("species==@species_var")
```

```python
#grouping by species 
df.groupby("species")["sepal_width"].mean()
```

```python
df.groupby("species").agg(
    avg_sepal_length=("sepal_length","mean"),
    max_petal_width=("petal_width","max"),
    std_of_sepalWidth=("sepal_width","std")
)
```

```python
df["petal_ratio"]=df["petal_length"]/df["petal_width"]
df.groupby("species")["petal_ratio"].mean()
df.groupby("species").agg(
    petal_ratio =("petal_ratio","mean")
)
```
